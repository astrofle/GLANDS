Example of how to calibrate drift scan data using the GLANDS Python tools.

```python
# Set up some python magic for the notebook.
%matplotlib widget
%load_ext autoreload
%autoreload 2
```

```python
# Import packages.
import sys
import matplotlib.pyplot as plt

from glands import datared
```

```python
# This is where the data lives.
#path = "/home/scratch/psalas/projects/DriftScan/data"
path = "/home/sdfits/"
projid = "AGBT22B_234_02" # The project/session identifier
vbank = "B" # Which VEGAS bank to work with.
```

```python
# Use the information from the previous cell to define the filename.
fitsfile = f"{path}/{projid}/{projid}.raw.vegas/{projid}.raw.vegas.{vbank}.fits"
```

```python
# Load the contents of the SDFITS file.
table = datared.load_sdfits(fitsfile)
```

```python
# Calibrate scan 1, polarization 0 (YY).
freq, cal_data = datared.calibrate_scan(table, 1, plnum=0)
```

```python
# Plot the time average of the scan.
fig = plt.figure(dpi=150)
ax1 = fig.add_subplot(111)
ax1.plot(freq.to("MHz"), cal_data.mean(axis=0))
ax1.set_xlabel("Frequency (MHz)")
ax1.set_ylabel("Antenna temperature (K)")
ax1.set_ylim(-1, 1);
```
